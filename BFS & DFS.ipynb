# Обход в ширину

```python
graph = {
    'A': ['B', 'C'],
    'B': ['A', 'D', 'E'],
    'C': ['A', 'E'],
    'D': ['B', 'F', 'G'],
    'E': ['B', 'C', 'F', 'G', 'H'],
    'F': ['D', 'E'],
    'G': ['D', 'E', 'H', 'I', 'J'],
    'H': ['E', 'G', 'J', 'K'],
    'I': ['G','J','L'],
    'J': ['I', 'L', 'G', 'H'],
    'K': ['H'],
    'L': ['J', 'I']
}
```

```python
def bfs(graph, start):
    Open = [start]  #нераскрытые вершины
    Closed = []  #раскрытые вершины
    result = []
    
    while Open:
        n = Open.pop(0)  #первый элемент из Open
        if n not in Closed:
            Closed.append(n)
            result.append(n)
            Open.extend(neighbor for neighbor in graph[n] if neighbor not in Closed and neighbor not in Open)
    
    return result
```

```python
start = 'A'
bfs = bfs(graph, start)
bfs
```

# Обход в глубину

```python
def dfs(graph, start):
    stack = [start] 
    visited = set()
    result = []

    while stack:
        u = stack.pop() 
        if u not in visited:
            visited.add(u) 
            result.append(u) 
            for neighbor in reversed(graph[u]):
                if neighbor not in visited:
                    stack.append(neighbor)

    return result
```

```python
start = 'A'
dfs = dfs(graph, start)
dfs
```
