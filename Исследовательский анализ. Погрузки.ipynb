# Исследование вероятности выполнения заявок логистической компанией

Логистическая компания занимается оказанием услуг по железнодорожной перевозке.  
Ежедневно в компанию поступают заявки на транспортировку товаров. Для каждой заявки известно направление перевозки (между какими станциями нужно осуществить транспортировку товаров), объем товаров, требующих перевозку, планируемая дата отправки, учетное направление перевозки (некоторые особенности перевозки). В течение жизненного цикла заявки (от ее регистрации в системе до её закрытия) с ней могут происходить различные изменения: изменение станций отправки и доставки, изменение величины прибыли, получаемой компанией в результате осуществления услуги по доставке, отмена заявки и т.д. 
В конце месяца из существующих заявок формируется заявочная компания, которая передается в отдел оптимизации, который составляет расписание движения составов, которое старается выполнить весь объем поступивших заявок.

**Глобальная задача:** используя инструменты машинного обучения предсказать вероятность выполнения заявки в заявочной компании. 

**Задача на сегодня**: используя инструменты визуализации данных и статистического анализа выявить признаки, которые оказывают влияние выполнение заявки.

Вам дана информацию о том, как формировалась заявочная компания с 01.01.2023 по 30.09.2023. Для каждой заявки представлены все изменения, вносимые оператором и результат выполнения заявки (выполнение, перенос, отмена).

**Описание исходных данных:**
 -   `Идентификатор погрузки` - уникальный номер погрузки в системе данных компании;
 -   `Дата редактирования` - когда вносились изменения в информацию о заявке на выполнение погрузки;   
 -   `Версия погрузки` - номер редакции;
 -   `Оператор` -  человек, осуществляющий редактирование заявки;
 -   `Дата отправления` - планируемая дата отправки погрузки;        
 -   `Дата отправления факт` - фактическая дата отправки погрузки;
 -   `Учетное направление` - тип перевозки; 
 -   `Кластер назначения`- из какого региона двигается состав;     
 -   `Кластер отправления` -  в какой регион двигается состав; 
 -   `Код назначения` - станция откуда нужно перевести товары;           
 -   `Код отправления` - станция, куда нужно перевезти товары;         
 -   `Количество вагонов` - количество вагонов в составе;      
 -   `Контрибуция` - прибыль, которую получит компания за перевозку;                         
 -   `Идентификатор переноса` - была ли перенесена погрука (-1 – заявка отменена, 0 – заявка выполнена, 1 – заявка перенесена).                 
     

**Целевой признак**: статус выполнения заявки (1 -- если заявка выполнена, 0 -- если заявка не выполнена (перенесена, отменена)).

```python
df['Статус заявки'] = df['Идентификатор переноса'].map({
    -1: 'Отменена',
    0: 'Выполнена',
    1: 'Перенесена'
})
```

Сразу подметим, что перенесенных заявок у нас нет

```python
def plot_stats(grouped_data, title):
    grouped_data.plot(kind='bar', stacked=True, figsize=(10, 6))
    plt.title(title)
    plt.xlabel('Категория')
    plt.ylabel('Количество заявок')
    plt.xticks(rotation=45, ha='right')
    plt.legend(title='Статус заявки', loc='upper right')
    plt.tight_layout()
    plt.show()

#по операторам
operator_stats = df.groupby(['Оператор', 'Статус заявки']).size().unstack(fill_value=0)
plot_stats(operator_stats, "Соотношение выполненных, отмененных и перенесенных заявок по операторам")

#по направлениям
direction_stats = df.groupby(['Направления', 'Статус заявки']).size().unstack(fill_value=0)
plot_stats(direction_stats, "Соотношение выполненных, отмененных и перенесенных заявок по направлениям")

#по учетным направлениям
route_type_stats = df.groupby(['Учетное направление', 'Статус заявки']).size().unstack(fill_value=0)
plot_stats(route_type_stats, "Соотношение выполненных, отмененных и перенесенных заявок по учетным направлениям")
```

График по направлениям визуализировать сложнее всего, так как данных слишком много. Остальные же вполне читабельны.
1. По операторам заявки мы уже анализировали, выводы приведены выше
2. По учетным направленям видим, что чаще всего отменяются заявки по импортным морским направлениям, ВРП и подсыл порожних КТК. Чаще всего выполняются при этом заявки также по этим же направлениям (по ним чаще происходят погрузки в целом)
3. Соотношение выполнынных заявок к отмененным позитивнее всего именно в ВРП перевозках.

# Создание новых признаков для исходного

```python
df1 = pd.read_excel( 'Orders_2023.xlsx')
df1['Дата отправления'] = pd.to_datetime(df1['Дата отправления'])
df1['Дата отправления факт'] = pd.to_datetime(df1['Дата отправления факт'])
df1['Дата редактирования'] = pd.to_datetime(df1['Дата редактирования'], format='mixed')
```

```python
#текущая длительность жизни заявки
df1['Текущая длительность жизни заявки'] = (df1['Дата редактирования'] - df1['Дата редактирования'].min()).dt.days

#близость к отправке
df1['Близость к отправке'] = (df1['Дата отправления'] - df1['Дата редактирования']).dt.days

#периодичность редакции
df1['Периодичность редакции'] = df1.groupby('Идентификатор погрузки')['Дата редактирования'].diff().dt.days

#афтершоки
df1['Афтершоки'] = (pd.Timestamp.now() - df1['Дата отправления факт']).dt.days

df1
```

# Доп.признаки для агрегированных данных

```python
#количество изменений до отправки
df['До отправки'] = df['Дата редактирования'] <= df['Дата отправления факт']
changes_before_sending = df[df['До отправки']].groupby('Идентификатор погрузки').size().rename('Изменения до отправки')

#количество изменений после отправки
changes_after_sending = df[~df['До отправки']].groupby('Идентификатор погрузки').size().rename('Изменения после отправки')

#количество переносов до отправки
transfers_before_sending = df[(df['Идентификатор переноса'] == 1) & (df['До отправки'])].groupby('Идентификатор погрузки').size().rename('Переносы до отправки')

#количество отмен до отправки
cancellations_before_sending = df[(df['Идентификатор переноса'] == -1) & (df['До отправки'])].groupby('Идентификатор погрузки').size().rename('Отмены до отправки')

#среднее количество времени, прошедшее между редакциями
df['Разница между редакциями'] = df.groupby('Идентификатор погрузки')['Дата редактирования'].diff().dt.days

#среднее количество времени между редакциями для каждой заявки
mean_time_between_edits = df.groupby('Идентификатор погрузки')['Разница между редакциями'].mean().rename('Среднее время между редакциями')

#длительность афтершока
df_last_edit = df.groupby('Идентификатор погрузки')['Дата редактирования'].max()
afterschock_duration = (df_last_edit - df.set_index('Идентификатор погрузки')['Дата отправления факт']).dt.days.rename('Длительность афтершока')

#все рассчитанные признаки в одном месте
aggregated_data = pd.DataFrame({
    'Изменения до отправки': changes_before_sending,
    'Изменения после отправки': changes_after_sending,
    'Переносы до отправки': transfers_before_sending,
    'Отмены до отправки': cancellations_before_sending,
    'Среднее время между редакциями': mean_time_between_edits,
    'Длительность афтершока': afterschock_duration
}).fillna(0)

aggregated_data
```
