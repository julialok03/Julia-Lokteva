```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, mean_squared_error
```

```python
np.random.seed(42)

#создадим случайные данные
n = 50
X = np.random.rand(n)
#зададим линейную зависимость в случайных данных
y_true = 2 + 3*X
#добавим шумы в данных
noise = np.random.normal(0, 0.5, n)
y = y_true + noise

plt.scatter(X, y)
```

```python
#создадим функцию для поиска ошибки
def mse(X, w1, w0, y):
    y_pred = w1 * X + w0
    return np.sum((y - y_pred) ** 2) / len(y_pred)
```

```python
e = 0.0001
coef_1 = 0
coef_0 = 0
learning_rate = 0.1
next_w1 = coef_1
next_w0 = coef_0
k=0

while True:
    k += 1
    cur_w1 = next_w1
    cur_w0 = next_w0
    y_pred = cur_w1 * X + cur_w0
    next_w0 = cur_w0 - learning_rate * (2/len(X)*np.sum((y - y_pred)) * (-1))
    next_w1 = cur_w1 - learning_rate * (2/len(X)*np.sum((y - y_pred) * (-X)))
    
    if (abs(cur_w1 - next_w1) <= e) and (abs(cur_w0 - next_w0) <= e):
        break
```

```python
coef_naklona = cur_w1
coef_sdviga = cur_w0
y_raschet = coef_naklona * X + coef_sdviga
```

```python
plt.scatter(X, y, color='blue', label='Исходные данные')
plt.plot(X, y_raschet, color='r', label='Регрессия')
plt.legend()
plt.show()
```
