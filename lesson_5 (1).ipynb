# Функция одной переменной

```python
from sympy import *

y, x = symbols('y x')

y = 3 * x ** 2
```

```python
x_values = np.linspace(-5, 5, 1000)
y_func = lambdify(x, y, 'numpy')
y_values = y_func(x_values)

plt.plot(x_values, y_values, label='y(x)')
```

```python
dif_x = diff(y, x)
dif_x
```

```python
x_1 = 5
lam = 0.1
res = []
x_0 = 0

while abs(x_0 - x_1) > 0.00000001:
    x_0 = x_1
    res.append([x_0, y.subs(x, x_0)])
    x_1 = x_0 - lam * dif_x.subs(x, x_0)
```

```python
import matplotlib.pyplot as plt

import numpy as np

x_values = np.linspace(-5, 5, 1000)
y_func = lambdify(x, y, 'numpy')
y_values = y_func(x_values)

plt.plot(x_values, y_values, label='y(x)')
plt.plot(*zip(*res), marker='o', linestyle='-', color='r', label='res')
plt.show()
```

# Функция двух переменных

```python
from sympy import *

f, y, x = symbols('f y x')

f = 3 * x ** 2 + 5 * y ** 2
```

```python
dif_x = diff(f, x)
dif_y = diff(f, y)
print(dif_x, ";", dif_y)
```

```python
x_1 = 5
lam = 0.1
res = []
x_0 = 0

while abs(x_0 - x_1) > 0.00000001:
    x_0 = x_1
    res.append([x_0, y, f.subs(x, x_0)])
    x_1 = x_0 - lam * dif_x.subs(x, x_0)
```

```python
x_1 = 5
lam = 0.1
res = []
x_0 = 0
y_0 = 0
y_1 = 10

while ((x_0 - x_1)**2+(y_0-y_1)**2)**0.5 > 0.00000001:
    x_0 = x_1
    y_0 = y_1
    res.append([x_0, y_0, f.subs(x, x_0).subs(y, y_0)])
    x_1 = x_0 - lam * dif_x.subs(x, x_0)
    y_1 = y_0 - lam * dif_y.subs(y, y_0)
```
